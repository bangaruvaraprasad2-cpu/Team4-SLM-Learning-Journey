3. Even or Odd Problem Statement: Take a number as input and check whether it is Even or Odd.

```python
num = int(input("Enter a number: "))

if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

Positive, Negative, or Zero Problem Statement: Take a number as input and determine whether it is: ● Positive ● Negative ● Zero

```python
num = float(input("Enter a number: "))

if num > 0:
    print("Positive")
elif num < 0:
    print("Negative")
else:
    print("Zero")
```

5. Largest of Two Numbers Problem Statement: Take two numbers as input and find the larger number.

```python
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

if num1 > num2:
    print(num1, "is larger")
elif num2 > num1:
    print(num2, "is larger")
else:
    print("Both numbers are equal")
```

Print Numbers from 1 to 100 Problem Statement: Use a loop to print numbers from 1 to 100.

```python
for i in range(1, 101):
    print(i)
```

7. Print Even Numbers from 1 to 100 Problem Statement: Print all even numbers between 1 and 100.

```python
for i in range(2, 101, 2):
    print(i, end=" ")
```
